```python
crossval_r2 = pd.DataFrame(index=range(2,4),columns = ["training","validation"])
```

```python
from sklearn.model_selection import KFold
```

```python
d = 5
kf = KFold(n_splits=10)
for d in range(2,25):
  r2_train_list = []
  r2_valid_list= []
  for train, valid in kf.split(X_train):
    regTree = ExtraTreesRegressor(max_depth=d)
    regTree.fit(X_train[train,:], y_train[train])
    R2_train = r2_score(y_train[train],regTree.predict(X_train[train,:]))
    R2_valid = r2_score(y_train[valid],regTree.predict(X_train[valid,:]))
    r2_train_list.append(R2_train)
    r2_valid_list.append(R2_valid)
  crossval_r2.loc[d,"training"] = np.mean(r2_train_list)
  crossval_r2.loc[d,"validation"] = np.mean(r2_valid_list)
```

```python
plt.scatter(crossval_r2.index,crossval_r2["training"],label="training")
plt.scatter(crossval_r2.index,crossval_r2["validation"],label="validation")
plt.xlabel("Komplexitas/Melyseg")
plt.ylabel("R2")
plt.legend()
plt.show()
```
